# Retail Sales Data â€“ Exploratory Data Analysis

This notebook demonstrates exploratory data analysis (EDA) on a small synthetic retail sales dataset containing both numerical and categorical variables.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'OrderID': range(1, 16),
    'ProductCategory': ['Electronics','Furniture','Clothing','Electronics','Electronics',
                        'Furniture','Electronics','Clothing','Furniture','Electronics',
                        'Clothing','Furniture','Electronics','Clothing','Furniture'],
    'Region': ['East','West','East','North','South','East','South','East','North','West',
               'South','East','North','West','South'],
    'UnitsSold': [5,2,7,3,8,4,6,5,3,7,2,4,6,5,3],
    'UnitPrice': [200,450,50,220,60,430,210,55,400,230,65,420,205,70,410],
    'CustomerRating': ['High','Medium','Low','High','Low','Medium','High','Low','Medium',
                       'High','Low','Medium','Medium','Low','Medium']
}

df = pd.DataFrame(data)
df['Revenue'] = df['UnitsSold'] * df['UnitPrice']
df
```

```python
df.info()
df.describe(include='all')
```

```python
for col in ['ProductCategory', 'Region', 'CustomerRating']:
    print(f"\n{col} counts:")
    print(df[col].value_counts())
```

```python
plt.figure()
plt.hist(df['UnitsSold'])
plt.title('Units Sold Distribution')
plt.show()

plt.figure()
plt.hist(df['UnitPrice'])
plt.title('Unit Price Distribution')
plt.show()

plt.figure()
plt.hist(df['Revenue'])
plt.title('Revenue Distribution')
plt.show()
```

```python
plt.figure()
plt.boxplot(df['UnitsSold'])
plt.title('Units Sold Boxplot')
plt.show()

plt.figure()
plt.boxplot(df['UnitPrice'])
plt.title('Unit Price Boxplot')
plt.show()

plt.figure()
plt.boxplot(df['Revenue'])
plt.title('Revenue Boxplot')
plt.show()
```

## Key Insights

- Electronics is the most frequent product category, indicating higher demand relative to other categories.
- Revenue varies substantially due to differences in price and volume.
- East and South regions show higher order counts.
- Customer ratings are predominantly Medium and Low, highlighting potential areas for service or product improvement.
